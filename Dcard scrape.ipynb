```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup as BS
import time
import matplotlib.pyplot as plt
```

```python
opt = Options()
```

```python
opt.add_argument('--headless')
opt.add_argument('--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.157 Safari/537.36')
#設定無頭模式
#設定 user-agent
```

```python
driver = webdriver.Chrome('chromedriver.exe', options=opt)
#開啟流覽器
```

```python
ua = driver.execute_script('return navigator.userAgent')
ua
#反回 user-agent 設定
```

```python
url = r'https://www.dcard.tw/f/job'
driver.get(url)
#前往頁面
```

```python
driver.save_screenshot('current.png')
```

```python
img = plt.imread('current.png')
plt.imshow(img)
```

```python
element = driver.find_element_by_class_name('PostList_footer_JYmTRV')
#定義查找的元素
```

```python
i = 1
while True:
    old_element = driver.find_elements_by_class_name('PostList_entry_1rq5Lf')
    print ('目前議題數：{}'.format(len(old_element)))
    print ('第 {} 回載入'.format(i))
    driver.execute_script('arguments[0].scrollIntoView(true)', element)
    time.sleep(2)
    new_element = driver.find_elements_by_class_name('PostList_entry_1rq5Lf')
    print ('目前議題數：{}'.format(len(new_element)))
    print ('==============')
    i+=1
    if len(new_element) == len(old_element):
        break
```
